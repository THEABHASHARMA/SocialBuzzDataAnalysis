```python
new_order = [
    'Content ID', 'Category', 'Category Type', 'Reaction Type',
    'Datetime', 'Sentiment', 'Score'
]
df = merged_df2[new_order]
```

```python
df.head()
```

```python
df['Category'] = df['Category'].str.replace('"', '', regex=False)
```

Top 5 performing categories

```python
category_scores = df.groupby('Category')['Score'].sum()
```

```python
top_categories = category_scores.sort_values(ascending=False).head(5)
```

```python
# data is mixed of capital and small letter categories making it duplicate
df['Category'] = df['Category'].str.lower()
```
